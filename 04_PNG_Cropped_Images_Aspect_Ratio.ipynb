# The PNG Cropped Images Aspect Ratio Notebook

##### This notebook reads all the windowed and cropped images.
##### Gets the width and height of all these images and calculates the aspect ratio.

## Import the Relevant Libraries

```python
# Allows for the access to the image directories
import os
```

```python
# Allows me to read the images to access the width and height 
from PIL import Image
```

```python
# Allows me to work with and manipulate the image array
import numpy as np
```

```python
# Allows for the plotting of the different image width and height sizes  
import matplotlib.pyplot as plt
```

## Defining the Directories

```python
# Cancer - cc view
dir_c_cc = "RSNA_PNG_Mammograms/cancer/cc/"

# Cancer - mlo view
dir_c_mlo = "RSNA_PNG_Mammograms/cancer/mlo/"


# No cancer - cc view
dir_nc_cc = "RSNA_PNG_Mammograms/no_cancer/cc/"

# No cancer - mlo view
dir_nc_mlo = "RSNA_PNG_Mammograms/no_cancer/mlo/"
```

## Create Lists of Images from the Directories 

```python
# Cancer - CC view
cancer_cc = os.listdir(dir_c_cc)

# Cancer - MLO view
cancer_mlo = os.listdir(dir_c_mlo)
```

```python
# No cancer - CC view
no_cancer_cc = os.listdir(dir_nc_cc)

# No cancer - MLO view
no_cancer_mlo = os.listdir(dir_nc_mlo)
```

## Get the Width and Height of the Images

```python
# Create empty list that will contain all the width diameters of the images
dim_x = []

# Create empty list that will contain all the height  diameters of the images
dim_y = [] 
```

```python
# Iterate through the image list in the specified directory 
def get_width_height(spes_lst, spes_dir):
    for image in spes_lst:
        # Define the source path
        image_path = spes_dir + image

        # Read the image 
        img = Image.open(image_path)
        
        # Obtain the width and height diameter
        width, height = img.size

        # Add them to the relevant list
        dim_x.append(width)        
        dim_y.append(height)
```

```python
# Cancer - CC view
get_width_height(cancer_cc, dir_c_cc)

# Cancer - MLO view
get_width_height(cancer_mlo, dir_c_mlo)
```

```python
# No cancer - CC view
get_width_height(no_cancer_cc, dir_nc_cc)

# No cancer - MLO view
get_width_height(no_cancer_mlo, dir_nc_mlo)
```

```python
# Make the plot
fig = plt.figure(figsize = (12, 4))

# Plotting
plt.scatter(dim_x, dim_y)

# Adding labels and title
plt.xlabel("Width")
plt.ylabel("Height")
plt.title("Width and Height Distribution of DICOM Images")

# Add gridlines for better visualization
plt.grid(True, axis = "y")  

# Show plot
plt.show()
```

## Determine Aspect Ratio and Resize Size

```python
# Calculate the mean width 
mean_width  = round(np.mean(dim_x), 0)
print(mean_width)
```

```python
# Calculate the mean height
mean_height  = round(np.mean(dim_y), 0)
print(mean_height)
```

```python
# Calculate the aspect ratio
aspect_ratio = mean_width / mean_height
print(aspect_ratio)
```
